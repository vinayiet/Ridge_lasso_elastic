```python
from sklearn.datasets import make_classification 
from sklearn.linear_model import LogisticRegression
X, y = make_classification(n_samples = 1000, n_features = 10, n_redundant = 5, n_informative = 5, n_classes =2, random_state = 1) 
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size = .30 , random_state = 1)
```

```python
model = LogisticRegression()
```

```python
model
```

```python
# Hyper parameter tuning 
# Grid search cv
from sklearn.model_selection import GridSearchCV
```

```python
params = {'penalty':('l1', 'l2','elasticnet'), 'C':[1,2, 10, 20, 30, 40]}
```

```python
classifier = LogisticRegression()
```

```python
clf = GridSearchCV(classifier, param_grid = params, cv = 5)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
model  = LogisticRegression( C = 1, penalty  = 'l2')
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.predict_proba(x_test)
```

```python
# Randomized search cv , It will take randomly some combination
```

```python
from sklearn.model_selection import RandomizedSearchCV

random_clf = RandomizedSearchCV(classifier, param_distributions = params, cv =5, n_iter = 10)
```

```python
random_clf
```

```python
random_clf.fit(X_train, y_train)
```

```python
random_clf.best_score_
```
