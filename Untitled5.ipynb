```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=2)
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y = pd.DataFrame(X)[1], hue = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel = 'linear')
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.coef_
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

params = {'C': [1, 2, 3, 10, 50], 'gamma': [1, 0.1, 0.3, 0.001, 0.003], 'kernel': ['linear']}
grid = GridSearchCV(SVC(), param_grid=params, cv=5, verbose=True)
grid.fit(X_train, y_train)
```
